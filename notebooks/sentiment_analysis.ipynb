## FlaxGPT2FewShot

It‚Äôs GPT-2 model. It was trained on the OSCAR dataset https://huggingface.co/flax-community/gpt2-base-thai

```python
flaxgpt2_model = FlaxGPT2FewShot('./flex-model1')
```

```python
flaxgpt2_model.train(
    data,
    logging_dir='./log_model_flex_1',
    num_train_epochs=300,
    train_size=0.9,
    batch_size=6,
    save_every_epochs=False
)
```

## GPTNeoFewShot

GPT-Neo is a transformer model designed using EleutherAI‚Äôs replication of the GPT-3 architecture. https://huggingface.co/EleutherAI

```python
gptneo_model = GPTNeoFewShot('./gptneo-model-1', model_name='gpt-neo') # try changed to model_name='thaigpt-next'
```

```python
gptneo_model.train(
    data,
    logging_dir='./log_model_neo_1',
    num_train_epochs=300,
    train_size=0.9,
    batch_size=2,
    save_every_epochs=False
)
```
