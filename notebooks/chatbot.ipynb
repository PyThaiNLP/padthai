```python
data = [
        """
        Q: สวัสดี\n
        A: สวัสดีจ้า\n
        Q: ทำอะไรอยู่\n
        A: กำลังนั่งเล่นจ้า แล้วคุณล่ะ\n
        Q: กำลังกินข้าวเหมือนกัน\n
        A: จ้า ^_^
        """,
        """
        Q: ร้องเพลงให้ฟังหน่อย\n
        A: ช้าง ช้าง ช้าง น้องเคยเห็นช้างรึเปล่า\n
        Q: เลือกเพลงได้ดีนี่\n
        A: ฮา ฮา คุณก็ลองบ้างสิ\n
        Q: ไม่ดีกว่าน่า\n
        A: โกงนี่น่า
        """,
        """
        Q: เครียดกับโควิดจริง ๆ ช่วงนี้\n
        A: พยายามทำใจให้สบายนะ :)
        """
]
```

```python
from padthai import *
```

```python
flaxgpt2_model = FlaxGPT2FewShot('./flex-model1')
```

```python
flaxgpt2_model.train(
    data,
    logging_dir='./log_model_flex',
    num_train_epochs=330,
    train_size=0.9,
    batch_size=8,
    save_every_epochs=False
)
```

```python
flaxgpt2_model.gen("""
Q: ร้องเพลงให้ฟังหน่อย\n
A:""",max_length=100)
```
